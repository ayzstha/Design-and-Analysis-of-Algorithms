# Activity 6: First Asymptotic Notation

The following code creates a plot demonstrating that $2x^2 = O(x^3)$ with $c=1$ and $n_0 = 2$.


```python
# matplotlib.pyplot is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. It also opens figures on your screen, and acts as the figure GUI manager.
import matplotlib.pyplot as plt
# numpy = numerical python...for arnge-Return evenly spaced values within a given interval.
import numpy as np

def f(x):
    return 2*x**2

def g(x):
    return x**3

n0 = 2
xmin = 0
xmax = 4

# numpy.linspace(start, stop, num) generates num numbers starting from start and ending at stop. 
#   The numbers are evenly spaced, meaning the difference between consecutive numbers is constant.
x = np.linspace(xmin, xmax, 50)

plt.plot(x, f(x), color = 'green', label = "$2x^2$")
plt.plot(x, g(x), color = 'blue', label = "$x^3$")
# Add a vertical line across the Axes, x position in data coordinates of vertical value (default is 0)
plt.axvline(x = n0, ymin = 0, ymax = 0.5, color = 'red', label = f'$n_0 = ${n0}')
# horizontal axis
plt.axhline(0, color = 'black')
# vertical axis
plt.axvline(0, color = 'black')



plt.title("Showing $2n^2 = O(n^3)$")
plt.xlim(xmin, xmax+1)
plt.ylim(0, 60)
plt.xlabel("x")
plt.ylabel("y")
plt.legend() # show labels
plt.show()
```

## Exercise 1.  $\Omega(g(n))$

Create a plot demonstrating that $\sqrt{n} = \Omega(\lg n)$ with $c=1$ and $n_0 = 16$.

```python
# matplotlib.pyplot is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. It also opens figures on your screen, and acts as the figure GUI manager.
import matplotlib.pyplot as plt
# hint: numpy has many useful built-in functions https://numpy.org/doc/stable/reference/routines.math.html
import numpy as np

def sqrt_n(n):
    return np.sqrt(n)

def log_n(n):
    return np.log2(n)

n0 = 16
xmin = 0
xmax = 25

# numpy.linspace(start, stop, num) generates num numbers starting from start and ending at stop. 
#   The numbers are evenly spaced, meaning the difference between consecutive numbers is constant.
x = np.linspace(xmin, xmax, 50)

plt.plot(x, sqrt_n(x), color = 'green', label = "$sqrt(n)$")
plt.plot(x, log_n(x), color = 'blue', label = "$c * log(n)$")
# Add a vertical line across the Axes, x position in data coordinates of vertical value (default is 0)
plt.axvline(x = n0, ymin = 0, ymax = 0.5, color = 'red', label = f'$n_0 = ${n0}')
# horizontal axis
plt.axhline(0, color = 'black')
# vertical axis
plt.axvline(0, color = 'black')



plt.title("Showing sqrt(n) = Ω(log(n))")
plt.xlim(xmin, xmax+1)
plt.ylim(0, 8)
plt.xlabel("x")
plt.ylabel("y")
plt.legend() # show labels
plt.show()

```

## Exercise 2.  $\Theta(g(n))$

Create a plot demonstrating that  $n^2/2 - 2n = \Theta(n^2)$ with $c_1=1/4, c_2 = 1/2$ and $n_0 = 8$.

```python
# matplotlib.pyplot is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. It also opens figures on your screen, and acts as the figure GUI manager.
import matplotlib.pyplot as plt
# hint: numpy has many useful built-in functions https://numpy.org/doc/stable/reference/routines.math.html
import numpy as np

def f(n):
    return (n**2) / 2 - 2 * n

def theta_n_squared(n):
    return n**2

n0 = 8
xmin = 0
xmax = 18
c1 = 1/4
c2 = 1/2

# numpy.linspace(start, stop, num) generates num numbers starting from start and ending at stop. 
#   The numbers are evenly spaced, meaning the difference between consecutive numbers is constant.
x = np.linspace(xmin, xmax, 50)

plt.plot(x, f(x), color = 'green', label = "$(n**2) / 2 - 2 * n$")
plt.plot(x, c1*theta_n_squared(x), color = 'blue', label = "$c1 * n**2$")
plt.plot(x, c2*theta_n_squared(x), color = 'black', label = "$c2 * n**2$")

# Add a vertical line across the Axes, x position in data coordinates of vertical value (default is 0)
plt.axvline(x = n0, ymin = 0, ymax = 0.5, color = 'red', label = f'$n_0 = ${n0}')
# horizontal axis
plt.axhline(0, color = 'black')
# vertical axis
plt.axvline(0, color = 'black')



plt.title("Showing (n**2) / 2 - 2 * n = theta(n^2)")
plt.xlim(xmin, xmax+1)
plt.ylim(0, 200)
plt.xlabel("x")
plt.ylabel("y")
plt.legend() # show labels
plt.show()

```
